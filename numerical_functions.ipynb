```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
# define x-axis values
x = np.linspace(-10, 10, 20)
# define function
y = x**2

# clear current plot
plt.clf()

plt.title("y=x^2")
plt.ylabel("y")
plt.xlabel("x")
plt.plot(x,y,'.')
plt.show()
```

```python
# we sample a sin(x) function
dx = np.pi/10
x = np.arange(0,2*np.pi,np.pi/10)

# we calculate the derivative, with np.gradient
plt.plot(x, np.gradient(np.sin(x), dx), '-*', label='approx')

# we compare it with the exact first derivative, i.e. cos(x)
plt.plot(x,np.cos(x), label='exact')
plt.legend()
```

```python
x = np.linspace(0,10,1000)
dx = x[1]-x[0]
y = x**2 + 1
dydx = np.gradient(y, dx, edge_order=2)
plt.plot(x, dydx, label='f\'(x)')
plt.legend()
```

### Prklad 1

S toleranciou 10e-10 njdite priblin hodnotu najvieho korea metdou linernej interpolcie - **metda tetv**.
Vypotajte odhad absoltnej chyby aproximcie.

```python
def plot_fn(f):
    x = np.linspace(-10, 10, 50)
    y = f(x)
    # clear current plot
    plt.clf()

    plt.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
    plt.axvline(x=0, c="black")
    plt.axhline(y=0, c="black")
    plt.ylabel("y")
    plt.xlabel("x")
    plt.plot(x,y)
    plt.show()

plot_function(f=lambda x: x**2 + 3*x)
```

```python
# https://ibug.io/blog/2019/05/mathematica-keygen/

import numpy as np
import math

def f1(x):
    return x**2+np.log(x)-10/x  # np.log is better than math.log

def f2(x):
    return x**5-3*x**2+1
```

```python
plot_function(f2)
```

```python
def tetivova_metoda(x=[1, 1.5], tolerate=10e-10, f=f2, r=10):
    for i in range(1, r+1):
        x_plus_1 = x[i] - f(x[i]) * ((x[i] - x[i-1]) / (f(x[i]) - f(x[i-1])))
        print(f'i: {i-1}  x(i):{x[i-1]}   |x(i+1) - x(i)|:{abs(x[i] - x[i-1])}')
        if abs(x[i] - x[i-1]) < tolerate:
            break
        x.append(x_plus_1)
    print('\n')

tetivova_metoda(x=[1, 1.5], f=f2, r=20)
tetivova_metoda(x=[-5, 5], f=f2)
tetivova_metoda(x=[-7.5, -5], f=f2, r=10)
tetivova_metoda(x=[-7.5, -5], f=f2, r=20)
```
