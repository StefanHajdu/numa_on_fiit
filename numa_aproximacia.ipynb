```python
import numpy as np
import matplotlib.pyplot as plt
import numa as nm

def generate_random_exp_data():
    np.random.seed(20210706)
    x = np.arange(10)
    a = 4.5
    b = 0.5
    c = 50
    y = a * np.exp(b * x) + c
    y = y + np.random.normal(scale=np.sqrt(np.max(y)), size=len(x))
    return x, y

def scatter_plot_data(x, y):
    ax = plt.axes()
    ax.scatter(x, y)
    ax.set_title('Example Data')
    ax.set_ylabel('y-Values')
    ax.set_ylim(0, 500)
    ax.set_xlabel('x-Values')

```
