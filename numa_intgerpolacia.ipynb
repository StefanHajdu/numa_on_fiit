### Interpolácia pomocu Newtonovho polynómu

```python
import numpy as np
import numa as nm
import sympy as sp
```

```python
x = np.array([0.142857, 0.285714, 0.428571, 0.571429, 0.714286, 0.857143, 1])
y = np.array([0, 1.38629, 3.29584, 5.54518, 8.04719, 10.7506, 13.6214])

polynom = nm.interpolacia_newtonov_polynom(x, y)
print(f'Newtonov polynom (zakladany tvar):\n\n{sp.expand(polynom)}\n')
nm.interpolacia_hodnota_bod(polynom, 0.3)
```

```python
x = np.array([0.142857, 0.285714, 0.428571, 0.571429]) #, 0.714286, 0.857143, 1])
y = np.array([0, 1.38629, 3.29584, 5.54518]) # 8.04719, 10.7506, 13.6214])

polynom = nm.interpolacia_newtonov_polynom(x, y)
nm.interpolacia_hodnota_bod(polynom, 0.3)
```
