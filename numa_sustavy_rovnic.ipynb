```python
import sympy as sp
import numa as nm
```

### Sustavy nelinearnych rovnic

Vzorovy priklad:

```python
_vars = {'x': 0, 'y': 0}
_funcs = ['2*x-y+(1/9)*exp(1)**(-x)-1', '-x+2*y+(1/9)*exp(-y)']

nm.plot_fn_system(_vars, _funcs, x_axis=(-5, 5, 100))
nm.plot_fn_system(_vars, _funcs, x_axis=(0, 1, 100))
nm.plot_fn_system(_vars, _funcs, x_axis=(-5, -1, 100))
```

```python
_vars = {'x': 0, 'y': 2}
nm.newton_sustava_rovnic(_vars, _funcs)

_vars = {'x': -2.5, 'y': -5}
nm.newton_sustava_rovnic(_vars, _funcs)

# body v blizkosti minima
# _vars = {'x': -3.5, 'y': -2.5}
# nm.newton_sustava_rovnic(_vars, _funcs)
```
